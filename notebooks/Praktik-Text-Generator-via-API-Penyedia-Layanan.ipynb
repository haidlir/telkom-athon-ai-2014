**Kelebihan**:
- Kemudahan penggunaan: Anda tidak perlu membangun infrastruktur Anda sendiri.

- Keamanan: Penyedia layanan biasanya memiliki langkah-langkah keamanan yang kuat untuk melindungi data Anda.

**Kekurangan**:
- Biaya: Penyedia layanan berbayar dapat menguras biaya Anda.
- Kontrol: Anda memiliki kontrol yang lebih sedikit atas model dan datanya.
- Privasi: Data Anda mungkin disimpan oleh penyedia layanan.

# Open-AI

```python
from google.colab import userdata
from openai import OpenAI

client = OpenAI(
    api_key=userdata.get("OAI_KEY")
)

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are an ML Engineer, skilled in explaining complex machine learning concepts with creative flair."},
    {"role": "user", "content": "Compose in Bahasa Indonesia, an illustration story that explains the concept of Large Language Model!"}
  ]
)

print(completion.choices[0].message.content)
```

```python
import anthropic
from google.colab import userdata

client = anthropic.Anthropic(
    api_key=userdata.get("ANT_KEY")
)

message = client.messages.create(
    model="claude-3-sonnet-20240229",
    max_tokens=1024,
    temperature=0.9,
    system="You are an ML Engineer, skilled in explaining complex machine learning concepts with creative flair.",
    messages=[
        {"role": "user", "content": "Compose in Bahasa Indonesia, an illustration story that explains the concept of Large Language Model!"}
    ]
)
print(message.content[0].text)
```
